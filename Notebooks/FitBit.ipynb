```python

activity.drop('SedentaryActiveDistance', axis=1, inplace=True)
activity.head()
```

```python
sleep.drop('logId', axis=1, inplace=True)
sleep.head()
```

```python
sleep['date'] = pd.to_datetime(sleep['date'])
sleep['adjusted_date'] = sleep['date'] - pd.Timedelta(hours=12)
sleep['sleep_day'] = sleep['adjusted_date'].dt.date
```

```python
asleep = sleep[sleep['value'] == 1]
restless = sleep[sleep['value'] == 2]
awake = sleep[sleep['value'] == 3]
```

```python
sleep_minutes = asleep.groupby(['Id','sleep_day']).size().reset_index(name='minutes_asleep')
restless_minutes = restless.groupby(['Id','sleep_day']).size().reset_index(name='minutes_restless')
awake_minutes = awake.groupby(['Id','sleep_day']).size().reset_index(name='minutes_awake')
```

```python
sleep_daily = sleep_minutes.merge(restless_minutes, on=['Id','sleep_day'], how='left')
sleep_daily = sleep_daily.merge(awake_minutes, on=['Id','sleep_day'], how='left')
sleep_daily.fillna(0, inplace=True)
```

```python
sleep_daily.head()
```

```python
sleep_daily['sleep_hours'] = sleep_daily['minutes_asleep'] / 60
sleep_daily['restless_hours'] = sleep_daily['minutes_restless'] / 60
sleep_daily['awake_hours'] = sleep_daily['minutes_awake'] / 60
```

```python
sleep_daily.head()
```

```python
sleep_daily['time_in_bed'] = (
    sleep_daily['minutes_asleep'] +
    sleep_daily['minutes_restless'] +
    sleep_daily['minutes_awake']) /60

sleep_daily['sleep_efficiency'] = sleep_daily['sleep_hours'] / sleep_daily['time_in_bed']
```

```python
sleep_daily.head()
```

```python
sleep = sleep_daily.copy()
sleep['sleep_category'] = pd.cut(
    sleep['sleep_hours'],
    bins=[0,5,7,9],
    labels=['low','moderate','high']
)
```

```python
activity['ActivityDate'] = pd.to_datetime(activity['ActivityDate'], format='%m/%d/%Y')
activity['ActivityDate'] = activity['ActivityDate'].dt.date
sleep['sleep_day'] = pd.to_datetime(sleep['sleep_day']).dt.date
```

```python
activity_clean = activity.groupby(['Id','ActivityDate'], as_index=False).agg({
    'TotalSteps': 'sum',
    'TotalDistance': 'sum',
    'Calories': 'sum',
    'VeryActiveMinutes': 'sum',
    'FairlyActiveMinutes': 'sum',
    'LightlyActiveMinutes': 'sum',
    'SedentaryMinutes': 'sum'
})
```

```python
fitbit = pd.merge(activity_clean, sleep, left_on=['Id','ActivityDate'], right_on=['Id','sleep_day'], how='inner')
fitbit.shape
```
